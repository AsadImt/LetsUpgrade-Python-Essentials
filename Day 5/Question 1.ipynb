```python
givenList = [1, 1, 5]

lst = []

lngth = int(input("\nEnter the length of the list : "))
j = 0;

#input-

for i in range (0,lngth):
    lst.append(int(input("\nEnter the number for the list : ")))
print()
print("List is : ", lst)

#process

for i in range(0,lngth):
    
    if( lst[i] == givenList[j] ):
        j += 1
        i += 1
    else:
        i += 1
if( j == 3 ):
    print ("\nIt's a match")
else:
    print ("\nIt's Gone")
```
