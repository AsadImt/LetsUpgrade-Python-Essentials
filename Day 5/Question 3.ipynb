```python
print('Make a Lambda function for capitalizing the whole sentence passed using arguments. And map all the sentences in the List, with the lambda functions')

givenList = ["hey! this is Asad","i am from Bihar"]
print("\nOriginal list : \n")
print(givenList)

result = list(map(lambda words: " ".join([word.capitalize() for word in words.split( )]) ,givenList))

print("\nCapitalized list is : \n")

print(result)
```
