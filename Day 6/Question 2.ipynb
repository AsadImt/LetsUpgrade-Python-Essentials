```python
import math
print('Question 2. \nFor this challenge,create a cone class that has two attributes: \n*R=Radius \n*h=Height \nAnd two methods: \n*Volume = Π * r2 = (h/3) \n*Surface area : base : Π * r2 , side : Π * r * √(r2 + h2) \nMake only one class with functions,as in where required import Math.')
pi = math.pi
class cone:
    def __init__(self,r,h): 
        self.r=r        
        self.h=h        
    def volume(self):        
        result = (1 / 3) * pi * self.r * self.r * self.h        
        print("\nVolume Of Cone is :",result)        
    def surfacearea(self):        
        result = pi * self.r * self.h + pi * self.r * self.r         
        print("\nSurface Area Of Cone is :",result)
radius = float(input("\nEnter the radius of cone : "))

height = float(input("\nEnter the height of cone : "))
c = cone( radius, height)
c.volume()

c.surfacearea()
```
